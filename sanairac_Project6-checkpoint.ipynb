# Zipf's Law

Noting the frequency of the distribution of words in a text has been a key aspect of statistical linguistics. The distribution of words in a text will follow a distinct mathematical pattern as dictated by Zipf's Law: when words are ranked by order of popularity, the most popular word will occur twice as often as the second most popular word, thrice as often as the third most popular word and so forth. Zipf's Law describes the phenomenon of exponential distribution and it can be observed in the population dynamics of cities and distribution of income, as well word frequencies. Zipf's law exhibits a universal behaviour in which frequency decreases rapidly with rank. We can write Zipf's law in the following equation: $logf = m.logr + b$, where $logf$ and $logr$ are linear with $m$ and $b$ represent the slope and intercept. The formula for the power-law is derived from this formula by introducing exponentials: 
$$e^{logf} = e^{m.logr + b}$$
$$e^{mlogr}e^{b} = (e^{logr})^{m}.e^{b}$$
$$f = r^{m}e^{b}$$
$$f = kr^{m}$$

The above derived equation $f = kr^{m}$ represents the power law distribution, or the "power law relationship". $f$ and $r$ are the variables of interest, $k$ is the constant and $m$ is the law's exponent. An inverse relationship like $f = r^{-1}$ is also a power law, because the change in one quantity will result in the negative change of another. Thus, this particular power law is able to explain how a small number of items can cluster themselves at the top of a distribution curve with a large number of occurences scattered at the botton. 

For this particular project, Zipf's Law will be used to study the distribution of word frequencies between different authors and different languages. We will use our own texts to test assumptions and better understand how language follows Zipf's Law. We will measure the frequency of a word relative to its rank. For the study we will extract text from two books with separate viewpoints but similar backgrounds: 'The Anthem' by Ayn Rand and 'The Communist Manifesto' by Karl Marx and Friedrich Engels. We will do the same for the a book translated into two different languages: 'Anna Karenina' translated into Finnish and Dutch. We will plot the frequency and rank against each other on a logarithmic axes, in an attempt to show a linear relationship which will indicate that these two quantities have a power law distribution.  

We will begin by importing the modules and packages that will define and implement the functions, classes, or variables that will be used over the course of this report:

```python
from numpy import *
import matplotlib.pyplot as plt
%matplotlib inline
from numpy import *
```

The following subplots will demonstrate the dependency between the rank and frequency. In order to obtain these plots we will first test a method to obtain the neccessary string from our word files and plot them accordingly. It will also give us an opportunity to observe a close comparison between the distribution of words for our selected novels.

```python
def BookPlot(text):

    file = open(text)
    book = file.read()  # read entire file into a single text string
    file.close()        # close the file

    book = book.lower() # change any upper case text to lower case

    punc = ',.:;?!()[]"\'@#_&^%~/<>|+=`'
    for c in punc: book = book.replace(c,'') #remove all punction
    lines = book.split() 

    wc = {}  
    for word in lines:
        if word not in wc: wc[word] = 0
        wc[word] += 1


    wcl = list(wc.items())                 # generate list of words
    wcl = sorted(wcl,key=lambda x:x[1],reverse=True)  # lambda construct for defining a function inline
    rank = arange(len(wcl)) + 1
    freqs = [item[1] for item in wcl]
    plt.xlabel('rank')
    plt.ylabel('frequency')
    plt.loglog(rank,freqs)
    #plt.semilo
```

Having defined a function neccessary for plotting for rank against frequency, we will proceed to graph the two subplots:

```python
plt.figure(figsize=(20,5))  # setting graph size
plt.subplot(1,2,1)
BookPlot('pg1250.txt');     # The Anthem by Ayn Rand
BookPlot('Marx.txt');       # The Communist Manifesto by Engels & Marx
plt.legend(['The Anthem','The Communist Manifesto']);

plt.subplot(1,2,2)
BookPlot('AnnaKFinnish.txt');  # Anna Karenina in Finnish
BookPlot('AnnaKDutch.txt');    # Anna Karenina in Dutch
plt.legend(['Anna Karenina(Finnish)','Anna Karenina(Dutch)']);
```

The above subplots draw comparisons between two seemingly similar, yet very different texts. The Anthem and the Communist Manifesto appear to have a similar linear distribution through the portion of the plot between the ranks $10^1$ and $10^2$. However, The Communist Manifesto has a higher frequency for words in its uppermost ranks, there is wider distribution for the words occuring less frequently and it has a more gradual gradient than The Anthem. The Anthem appears to have a steeper gradient than The Communiest Manifesto, because even though higher ranking words occur less frequently, there is a more rapid decline in word frequency overall. 

The relationship in word frequencies between the two versions of Anna Karenina are also fascinating when observed against the first subplot. The Finnish translation has a higher frequency for its highest ranking word than the Dutch translation and also has a wider distribution of frequencies for its word ranks overall, which is similar to the relationship between word and rank observed for The Communist Manifesto and The Anthem. However, the Finnish translation of Anna Karenina has a more gradual gradient for word rank and frequency in comparison to the Dutch translation, unlike what was observed for The Anthem and The Communist Manifesto. 

We will further study the differences in gradients using a line of best fit for each word frequency plot, a technique which we have studied and applied previously. If Zipf's law holds true for the equation $log(f) = m.log(r) + b$, the graph of $log(f)$ against $log(r)$ should be a straight line with slope close to $-1$.

```python
def LineofFit(text,title):

    file = open(text)
    book = file.read()  # same steps used in the function BookPlot
    file.close()
    
    book = book.lower()

    punc = ',.:;?!()[]"\'@#_&^%~/<>|+=`'
    for c in punc: book = book.replace(c,'')
    lines = book.split()

    wc = {}  
    for word in lines:
        if word not in wc: wc[word] = 0
        wc[word] += 1

    wcl = list(wc.items())        # generate list of words
    wcl = sorted(wcl,key=lambda x:x[1],reverse=True)  # lambda construct for defining a function inline
    rank = arange(len(wcl)) + 1
    freqs = [item[1] for item in wcl]
    
    def best2fit(x,y):            # definition used to obtain line of best fit
        xbar = sum(x)/len(x)      #mean of values in x
        ybar = sum(y)/len(y)      #mean of values in y
        n=len(x)

        ynum= sum([xi*yi for xi,yi in zip(x, y)]) - n * xbar * ybar #zip both arrays of x and y together 
        ydem= sum([xi**2 for xi in x]) - n * xbar**2                

        alpha = ynum/ydem          #slope of line
        beta = ybar - xbar*alpha   #point of intersection
        return alpha,beta
    
    x = log10(rank)
    y = log10(freqs)               # producing a log-log plot

    #plt.plot(rank,freqs)
    alpha,beta = best2fit(x,y)
    plt.plot(x,y)
    yfit = [beta + alpha * xi for xi in x]              #line of best fit
    plt.plot(x, yfit, 'r', alpha= 0.7);                 #plotting the line of best fit through the values of x
    plt.xlabel('rank')
    plt.ylabel('frequency')
    plt.title(title);
    plt.grid(True)
    print('The gradient of the line of best fit for',title,'is',alpha)
```

```python
plt.figure(figsize=(20,5))
plt.subplot(1,2,1)
LineofFit('pg1250.txt', "The Anthem by Ayn Rand")
#plt.figtext(.15,.8, ftext, fontsize=11);
plt.subplot(1,2,2)
LineofFit('Manifesto.txt', "The Communist Maifesto by Engels & Marx")
```

The the line of best fit for both graphs has a gradient surprisingly close to $-1$. For The Anthem we can observe through the line of best fit that there is a large cluster of words in the lower frequency range, with a small number of words occuring at large frequencies. The Communist Manifesto has a more gradual decline in frequnecy with the line of best fit running through most of the points. Clearly the log-log plots are roughly linear, confirming a Zipf distribution for both, although the slope of $-1.147$ and $-0.960$ are a little lower and a little larger than the Zipfian ideal of $-1$.

```python
plt.figure(figsize=(20,5))
plt.subplot(1,2,1)
LineofFit('AnnaKFinnish.txt', "Anna Karenina in Finnish")
#plt.figtext(.15,.8, ftext, fontsize=11);
plt.subplot(1,2,2)
LineofFit('AnnaKDutch.txt', "Anna Karenina in Dutch")
```

Both the Finnish and Dutch translations of Anna Karenina also have a word frequency gradient of close to $-1$. At $-0.9388$, the Finnish translation is surprisingly less linear than the Dutch translation which has a gradient of $-1.199$. Even though it appears the Dutch translation has few word ranks of higher frequency, it has a greater number of word ranks which fall within a medium or low frequency bracket. 

The existence of the "power law relationship" is quite obvious through these graphs, with a gradient of $-1$ indicating the most common words rapidly decline in frequency as we move down each plotted line. We have observed the presence Zipf's Law in the distribution of word frequencies, but just to confirm this relationship we will attempt to observe the law in another distribution set: word length against word frequency.

```python
def Wordfreq(text, title):
    file = open(text)
    book = file.read()  # same steps used in BookPlot
    file.close()

    book = book.lower()

    punc = ',.:;?!()[]"\'@#_&^%~/<>|+=`'
    for c in punc: book = book.replace(c,'')
    lines = book.split()

    wc = {}  
    for word in lines:
        if word not in wc: wc[word] = 0
        wc[word] += 1

    wcl = list(wc.items())                   # generate list 
    wcl = sorted(wcl,key=lambda x:x[1],reverse=True) 
    length = [len(w[0])for w in wcl]         # measure for the length of each word
    freqs = [item[1] for item in wcl]        # obtain the frequncy of respective word

    plt.scatter(length,log10(freqs),alpha=0.1)
    plt.xlabel('word length')
    plt.ylabel('log word frequency')
    plt.savefig('foo.png')
    plt.title(title)
```

```python
plt.figure(figsize=(20,5))
plt.subplot(1,2,1)
Wordfreq('pg1250.txt', "The Anthem by Ayn Rand")
#plt.figtext(.15,.8, ftext, fontsize=11);
plt.subplot(1,2,2)
Wordfreq('Manifesto.txt', "The Communist Maifesto by Engels & Marx")
```

Here we can observe the power law distribution component more clearly. By measuring the frequencies of word lengths on logarithmic scales, we can see the characteristic straight-line form of the power law distribution. The power-law distribution causes each progressive bin in the graphs to contain fewer words, if any. This change is more rapid and is more visible in The Anthem than it is for The Communist Maniesto. 

```python
plt.figure(figsize=(20,5))
plt.subplot(1,2,1)
Wordfreq('AnnaKFinnish.txt', "Anna Karenina in Finnish")
#plt.figtext(.15,.8, ftext, fontsize=11);
plt.subplot(1,2,2)
Wordfreq('AnnaKDutch.txt', "Anna Karenina in Dutch")
```

Repeating the above procedure for the Finnish and Dutch translations of Anna Karenina, we can see similar results, except higher levels of density for words falling in the bins between $5$ and $10$.  

## Conclusion

In conclusion, the existence of the power-law relationship stated in Zipf's Law is clearly visible through the linearity of our word frequency distribution graphs. With the line of best fit producing a gradient of almost $-1$, we will be able to conclude that for example, the $40$th most common word is half as frequent as the $20$th most common word, by using the equation for the power-law relationship:

$$\frac{f_{40}}{f_{20}} = \frac{k.40^{(-1)}}{k.20^{-1}} = \frac{20}{40} = \frac{1}{2}$$

The word frequency distributions oberved in The Anthem, The Communist Manifesto and the translations of Anna Karenina only serve to display a basic understaning of a more complex system that defines a universal phenomenon. The frequency dependance on rank are reflective of the exponential model dicated by Zipf's Law and can be observed discretely through the distribution of word length frequencies in the given texts.
